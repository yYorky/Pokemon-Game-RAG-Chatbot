```python
# Import necessary libraries
import pandas as pd
import numpy as np
from datasets import Dataset
from ragas.metrics import AnswerSimilarity, AnswerCorrectness
from ragas.evaluation import evaluate
import os
from dotenv import load_dotenv

load_dotenv()
os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY")
```

```python
# Load the CSV file
csv_file_path = 'C:/Users/York Yong/OneDrive - Singapore Management University/Desktop/Pokemon Game RAG Chatbot/Basline LLM evaluation/Eval_ChatGPT4o.csv'

try:
    data = pd.read_csv(csv_file_path, encoding='utf-8')
except UnicodeDecodeError:
    data = pd.read_csv(csv_file_path, encoding='latin1')
```

```python
# Define the metrics to be evaluated
metrics = [
    AnswerSimilarity(), 
    AnswerCorrectness()
]

# Initialize columns for results
data['answer_similarity'] = None
data['answer_correctness'] = None

# Iterate over each row to evaluate the metrics individually
for idx, row in data.iterrows():
    evaluation_data = Dataset.from_dict({
        'question': [row['question']],
        'ground_truth': [row['ground_truth']],
        'answer': [row['answer']]
    })
    
    evaluation_result = evaluate(evaluation_data, metrics=metrics)
    
    # Store results back in the DataFrame
    data.at[idx, 'answer_similarity'] = evaluation_result['answer_similarity']
    data.at[idx, 'answer_correctness'] = evaluation_result['answer_correctness']

```

```python
# Display the evaluation results
result_df = data[['question', 'ground_truth', 'answer', 'answer_similarity', 'answer_correctness']]
result_df
```

```python
# Print the mean values of answer_similarity and answer_correctness
print("Mean Answer Similarity:", data['answer_similarity'].replace(np.nan, 0).mean())
print("Mean Answer Correctness:", data['answer_correctness'].replace(np.nan, 0).mean())
```
